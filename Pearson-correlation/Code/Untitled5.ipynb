```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('userSimilarity.csv')
```

```python
df.head()
```

```python
df = df.fillna(0)
```

```python
pcaDataset = []
```

```python
#pca = PCA(n_components=5)
#model = pca.fit(df).transform(df)

for i in range(2,8):
    pca = PCA(n_components=i)
    model = pca.fit(df).transform(df)
    pcaDataset.append(model)
```

```python
print(len(pcaDataset))
```

```python
# Plot scatter plots in each subplot
fig, axs = plt.subplots(3, 2, figsize=(10, 12))

# Flatten the axs array for easier indexing
axs = axs.flatten()

# Iterate over each subplot and plot the scatter plot
for i, ax in enumerate(axs):
    x = pcaDataset[i]
    ax.scatter(x[:,0], x[:,1], s=0.3, c='blue', alpha=0.5)
    ax.set_title(f'n_components = {i+2}')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
print(sum(pcaDataset[1][:,0]-pcaDataset[0][:,0]))
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv('masterDf.csv')
df = df.fillna(0)

# Assuming you have your dataset stored in a NumPy array
dataset = df

# Create a KMeans object with the desired number of clusters (k)
kmeans = KMeans(n_clusters=2)

# Fit the model to your dataset
kmeans.fit(dataset)

# Retrieve the cluster labels assigned to each data point
labels = kmeans.labels_

# Retrieve the cluster centers
centroids = kmeans.cluster_centers_

```

```python
## Plot the dataset and cluster assignments
plt.figure()
plt.scatter(dataset, dataset, c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-means Clustering')
plt.show()
```
