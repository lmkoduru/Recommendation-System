```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('userSimilarity.csv')
```

```python
df.head()
```

```python
userSimilarityMatrix = df
userSimilarityMatrix = userSimilarityMatrix.set_index('user')
# userSimilarityMatrix.head()

examinedUserIndex = int(userSimilarityMatrix.columns[0])
userSimilarityMatrix.drop(index=examinedUserIndex, inplace=True)
userSimilarityMatrix.head()
```

```python
# Target number of similar users
n = 10

# Get top n similar users
#userSimilarityMatrix[0]
similarUsers = userSimilarityMatrix[str(examinedUserIndex)].nlargest(n).index.tolist()
values = userSimilarityMatrix[str(examinedUserIndex)].nlargest(n).tolist()

# Print out top n similar users
print(f'The similar users near user {examinedUserIndex} are', similarUsers)
print(f'The similarity values for user {examinedUserIndex} are', values)


```

```python
masterDf = pd.read_csv('masterDf.csv')
masterDf.head()
```

```python
masterDf1 = masterDf.set_index('user')
masterDf1.head()
```

```python
print(masterDf1.shape[0])
```

```python
masterDfNorm = masterDf1.subtract(masterDf1.mean(axis=1), axis = 'rows')
masterDfNorm.head(10)
```

```python
print(masterDfNorm.shape[0])
```

```python
# Movies that the target user has watched
picked_userid_watched = masterDfNorm[masterDfNorm.index == examinedUserIndex].dropna(axis=1, how='all')
picked_userid_watched
```

```python
masterDfNorm.head()
```

```python
similar_user_movies = masterDfNorm[masterDfNorm.index.isin(similarUsers)].dropna(axis=1, how='all')
similar_user_movies
```

```python
# Remove the watched movie from the movie list
similar_user_movies.drop(picked_userid_watched.columns,axis=1, inplace=True, errors='ignore')
# Take a look at the data
similar_user_movies
```

```python
# A dictionary to store item scores
similarUsersDict = dict(zip(similarUsers, values))

item_score = {}
# Loop through items
for i in similar_user_movies.columns:
    
  # Get the ratings for movie i
  movie_rating = similar_user_movies[i]
  # Create a variable to store the score
  total = 0
  # Create a variable to store the number of scores
  count = 0
  
 # Loop through similar users
  for u in similarUsersDict.keys():
    # If the movie has rating
    if pd.isna(movie_rating[u]) == False:
      # Score is the sum of user similarity score multiply by the movie rating
      score = similarUsersDict[u] * movie_rating[u]
      # Add the score to the total score for the movie so far
      total += score
      # Add 1 to the count
      count +=1
  # Get the average score for the item
  item_score[i] = total / count
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['movie', 'movie_score'])
    
# Sort the movies by score
ranked_item_score = item_score.sort_values(by='movie_score', ascending=False)
# Select top m movies
m = 10
ranked_item_score.head(m)

```
