```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
df = pd.read_csv('allmovies_cleaned.csv')
```

```python
df.head(10)
```

```python
users = df['user'].nunique()
movies = df['movie'].nunique()
print(str(users) + ' unique users in this dataset')
print(str(movies) + ' unique movies in this dataset')
```

```python
user_ratings = df.groupby(['user', 'movie'])['rating'].mean().astype(int)
print(user_ratings)
```

```python
masterDf = user_ratings.unstack(level='movie')
```
