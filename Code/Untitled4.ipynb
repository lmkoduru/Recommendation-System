```python
matrix_norm = df.subtract(df.mean(axis=1), axis = 'rows')
matrix_norm.head()
```

```python
column_counts = matrix_norm.count()
top_50_columns = column_counts.nlargest(50).index
df_filtered = matrix_norm[top_50_columns]
df_filtered.head()
```

```python
simpleMatrixNorm = matrix_norm.dropna(thresh=60)
len(simpleMatrixNorm)
```

# User-User Similarity Score

```python
user_similarity = simpleMatrixNorm.T.corr()
user_similarity.head()
```

```python
user_similarity.to_csv('./userSimilarity.csv')
```
